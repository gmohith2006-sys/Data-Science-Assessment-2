**Section A – Interview Theory**

Q1. A Series is a one-dimensional labeled array that stores data in the form of a single column whereas a DataFrame is a two dimensional table consisting of multiple Series sharing the same index. Operations like .head() function work on both while operations such as selecting multiple columns (df[['column_1','column_2']]) work only on a DataFrame.

Q2. inplace = True changes/modifies the original object instead of returning the new one. Pandas discourages its use because it often gives a little performance benefit and makes code harder to debug. I would avoid it when dropping columns eg: use df = df.drop('Salary', axis=1) which is an easier method instead of inplace=True

Q3. The .fillna() function replaces the specific missing values with a chosen value while the .dropna() function removes the rows or columns containing missing values. A data analyst uses .fillna() function when missing data should be filled such as replacing missing salaries with the median. And .dropna() function is useful when only a few incomplete records exist and removing them would not affect the analysis.

Concrete Situation:
- .fillna() : Analyzing the daily retail store sales.
- .dropna() : Predicting the credit risk or loan defaults where the target variable is missing.

Q4. The pd.to_numeric(col, errors='coerce') function silences the exceptions by converting the unparseable values like strings or garbage data into standard NaN values whereas the pd.to_numeric(col) defaults to errors='raise', which crashes your program with a ValueError the moment it encounters data then it cannot translate into a number. On real world messy datasets I would use errors='coerce' because it allows cleaning and analysis to continue without any difficulties.

Q5. The .loc selects the rows and columns using labels whereas the .iloc uses an integer positions. Example: df.loc['Age'] selects the row labeled 'Age' whereas df.iloc[0] selects the first row. If the DataFrame index is [8, 9, 10] then df.loc[8] returns the row labeled 8, while df.iloc[0] returns the first row which may have label 8 and if the index is reordered they can return different rows.

Q6. The .map() function is mainly used to transform the values using a dictionary, function, or another Series. The .apply() function applies a custom function to each element for more complex operations. The .replace() function substitutes specific values directly and is the best for simple value replacements without calculations.

1. Reach for the .map() function when categorizing or translating a full column.
2. Reach for the .replace() function when cleaning up specific typos or using Regex.
3. Reach for the .apply() function when evaluating complex Python logic.

Q7. The SettingWithCopyWarning occurs when modifying a slice of a DataFrame instead of the original DataFrame which makes the result uncertain. Example: df[df['Salary'] > 20000]['Salary'] = 30000 and the way to fix it is: df.loc[df['Salary'] > 20000, 'Salary'] = 30000

Q8. The pd.cut() function creates the bins with fixed value ranges while the pd.qcut() function creates the bins with approximately equal numbers of observations. For annual income data I would use the pd.qcut() function because it evenly distributes customers into Low, Mid, and High earning groups making the comparisons more balanced.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("indian_startups_fundingDADS.csv") # Load the dataset
# 1. Rename columns to snake_case
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
# 2. Clean startup_name and city
df["startup_name"] = (
    df["startup_name"]
      .astype(str)
      .str.strip()
      .str.title()
)
df["city"] = (
    df["city"]
      .astype(str)
      .str.strip()
      .str.title()
)
# 3. Standardise city aliases
city_mapping = {
    "Bangalore": "Bengaluru",
    "Blr": "Bengaluru",
    "Bombay": "Mumbai",
    "Gurgaon": "Gurugram",
    "Delhi": "New Delhi",
    "Calcutta": "Kolkata",
    "Madras": "Chennai"
}
df["city"] = df["city"].replace(city_mapping)
# 4. Standardise industry variants
industry_mapping = {

    "Fin-Tech": "Fintech",
    "FINTECH": "Fintech",
    "Fintech": "Fintech",

    "EdTech": "Edtech",
    "Edtech": "Edtech",

    "Health-Tech": "HealthTech",
    "Healthtech": "HealthTech",

    "Food Tech": "FoodTech",

    "Travel-Tech": "TravelTech",

    "Electric Vehicles": "Electric Vehicles",
    "Electric-Vehicles": "Electric Vehicles",

    "AI ML": "AI & ML",

    "SAAS": "SaaS",
    "Saas": "SaaS",
    "saas": "SaaS",

    "PROPTECH": "PropTech",
    "Prop-Tech": "PropTech",

    "Ecommerce": "Ecommerce",
    "ecommerce": "Ecommerce"
}
df["industry"] = df["industry"].replace(industry_mapping)
# 4(b). Standardise funding stage variants
stage_mapping = {

    "SEED": "Seed",
    "Seed": "Seed",
    "PRE-SEED": "Pre-Seed",

    "Pre-A": "Pre-Series A",
    "PRE-SERIES A": "Pre-Series A",
    "Pre-Series A": "Pre-Series A",

    "Series A": "Series A",
    "Series B": "Series B",
    "series_b": "Series B",

    "Series C": "Series C",
    "SERIES C": "Series C",

    "Series D": "Series D",

    "Series E": "Series E",
    "SERIES E": "Series E",
    "series e": "Series E",

    "IPO": "IPO",
    "ipo": "IPO",

    "Pre-IPO": "Pre-IPO",

    "BRIDGE": "Bridge",
    "bridge": "Bridge"
}
df["funding_stage"] = df["funding_stage"].replace(stage_mapping)
# 5. Convert Amount to numeric
df["amount"] = (
    df["amount"]
      .astype(str)
      .str.replace("₹", "", regex=False)
      .str.replace("INR", "", regex=False)
      .str.replace("Cr", "", regex=False)
      .str.replace("Crore", "", regex=False)
      .str.replace(",", "", regex=False)
      .str.strip()
      .replace("Undisclosed", np.nan)
)
df["amount"] = pd.to_numeric(df["amount"], errors="coerce")
# 6. Convert funding_date to datetime
df["funding_date"] = pd.to_datetime(
    df["funding_date"],
    errors="coerce",
    dayfirst=True
)
# Convert founded_year to numeric
df["founded_year"] = pd.to_numeric(
    df["founded_year"],
    errors="coerce"
)
# 7. Remove duplicate rows
df = df.drop_duplicates()
# Verify cleaned dataset
print(df.head())
print("\nShape :", df.shape)
print("\nColumn Names")
print(df.columns.tolist())
print("\nData Types")
print(df.dtypes)
```

**Aggregation & Grouping (Q33 to Q37)**

```python
# Q33.
top_5 = (
    df.groupby("industry")["amount"]
      .mean()
      .sort_values(ascending=False)
      .head(5)
)
print(top_5)
```

```python
# Q34.
startup_4 = (
    df.groupby("city")
      .agg({
          "startup_name": "count",
          "amount": ["sum", "mean"]
      })
)
startup_4.columns = ["startup_count", "total_funding", "average_funding"]
startup_4 = startup_4.sort_values(
    by="startup_count",
    ascending=False
)
print(startup_4.head(10))
```

```python
# Q35.
industry_stats = (
    df.groupby("industry")
      .agg(
          count=("startup_name", "count"),
          average_funding=("amount", "mean")
      )
)
industry_stats = industry_stats[
    industry_stats["count"] >= 15
]
industry_stats = industry_stats.sort_values(
    by="average_funding",
    ascending=False
)
print(industry_stats.head(1))
```

```python
# Q36.
top_5 = (
    df["industry"]
      .value_counts()
      .head(5)
      .index
)
cross = pd.crosstab(
    df[df["industry"].isin(top_5)]["industry"],
    df[df["industry"].isin(top_5)]["funding_tier"]
)
print(cross)
```
